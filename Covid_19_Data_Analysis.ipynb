```python
vaccine_df = pd.read_csv('/content/drive/MyDrive/Data_Development/Covid-19-Data-Analysis-Project-Using-Python-And-Tableau-main/Covid-19-Data-Analysis-Project-Using-Python-And-Tableau-main/covid_vaccine_statewise.csv')
```

```python
covid_df.drop(['Sno','Time','ConfirmedIndianNational','ConfirmedForeignNational'],inplace = True,axis=1)
# we use axis = 1 as we want to drop the columns and inplace = true because we are saving the changes in same dataset
```

```python
covid_df.head()
```

```python
# changing the date format to day-month-year

covid_df["Date"] = pd.to_datetime(covid_df["Date"],format = "%Y-%m-%d")
```

```python
covid_df.info()
```

ACTIVE CASES


```python
# Active cases
covid_df["Active_Cases"] = covid_df["Confirmed"] - ( covid_df["Cured"] + covid_df["Deaths"] )
```

```python
covid_df.dropna(how='any',axis=0,inplace=True)
# removing all the null values by columns
```

```python
covid_df.tail()
```

```python
# finding state_wise cured , deaths , confirmed Cases
statewise = pd.pivot_table(covid_df ,values =["Cured" , "Deaths" , "Confirmed"],index = "State/UnionTerritory",aggfunc = max)
```

```python
print(statewise)
```

```python
statewise["Recovery Rate"] = (statewise["Cured"]*100 / statewise["Confirmed"])
statewise["Death Rate"] = (statewise["Deaths"]*100 / statewise["Confirmed"])

statewise = statewise.sort_values(by="Confirmed",ascending = False)

statewise.style.background_gradient(cmap = "cubehelix")

```

Top 5 STATE with Highest Number Of Cases

```python
top_5_high_cases = covid_df.groupby(["State/UnionTerritory"] , as_index = False)["Confirmed"].max().sort_values( by = "Confirmed" , ascending = False).head()
sns.set(rc={'figure.figsize':(15,5)})
ax1  = sns.barplot(data = top_5_high_cases , x ="State/UnionTerritory" ,y = "Confirmed")

for bars in  ax1.containers:
  ax1.bar_label(bars)
```

TOP 5 state with highest number of Deaths

```python
top_5_high_death = covid_df.groupby(["State/UnionTerritory"], as_index = False)["Deaths"].max().sort_values(by = "Deaths",ascending = False).head()
sns.set({'figure.figsize':(15,6)})
ax2 = sns.barplot(data = top_5_high_death , x = "State/UnionTerritory" , y = "Deaths")

for bars in ax2.containers:
  ax2.bar_label(bars)
```

TOP 10 Active Cases


```python
top_10_active = covid_df.groupby(["State/UnionTerritory"],as_index = False)["Active_Cases"].max().sort_values(by="Active_Cases",ascending= False).head(10)
sns.set({'figure.figsize':(25,6)})
ax3 = sns.barplot(data = top_10_active , x = "State/UnionTerritory" , y = "Active_Cases")

for bars in ax3.containers:
  ax3.bar_label(bars)
```

Growth Trend Of Top 5 states

```python
sns.set(rc = {'figure.figsize':(25,5)})

ax  =  sns.lineplot(data = covid_df[covid_df["State/UnionTerritory"].isin(["Maharashtra","Karnataka","kerala","Uttar Pradesh","Tamil Nadu"])] , x = "Date" , y = "Active_Cases" , hue = "State/UnionTerritory" )
ax.set_title("Trend for Active Cases in Top 5 States", size = 16)
```

```python
# removing rows in which state is India
r_state = vaccine_df[vaccine_df.State!='India']
r_state.head()
```

```python
r_state.rename(columns ={"Total Individuals Vaccinated": "Total Vaccinated"},inplace = True)
```

Most Vaccinated State

```python
max_vc = r_state.groupby("State",as_index = False)["Total Vaccinated"].sum().sort_values(by = "Total Vaccinated", ascending = False).head()
```

```python
max_vc
```

```python
sns.set(rc = {'figure.figsize':(16,5)})
sns.barplot(data = max_vc , x = "State",y = "Total Vaccinated")
```

Least Vaccinated States

```python
min_vc = r_state.groupby("State",as_index = False)["Total Vaccinated"].sum().sort_values(by = "Total Vaccinated", ascending = False).tail()
sns.set({"figure.figsize":(16,5)})
sns.barplot(data = min_vc , x = "State" , y = "Total Vaccinated")
```
